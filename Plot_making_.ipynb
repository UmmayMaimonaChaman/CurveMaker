```python
import matplotlib.pyplot as plt

years = ["2020", "2021", "2022", "2023", "2024", "2025"]
data = {
    "Healthcare": [100, 250, 450, 612, 800, 950],
    "Medical Imaging": [50, 120, 200, 250, 350, 400],
    "IoT Health": [30, 80, 150, 200, 250, 300],
    "Smart Cities": [10, 30, 50, 60, 80, 100],
    "Biometrics": [5, 20, 40, 50, 60, 70],
    "Other": [40, 80, 150, 180, 220, 260]
}

bar_width = 0.13
x = range(len(years))
fig, ax = plt.subplots(figsize=(14, 6))

for i, (domain, vals) in enumerate(data.items()):
    ax.bar([p + i * bar_width for p in x], vals, width=bar_width, label=domain)

ax.set_xticks([p + 2.5 * bar_width for p in x])
ax.set_xticklabels(years)
ax.set_xlabel("Year")
ax.set_ylabel("No. of Publications")
ax.set_title("Federated Learning Papers by Domain (2020â€“2025)")
ax.legend()
plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt

# Years and data
years = [2020, 2021, 2022, 2023, 2024, 2025]
data = {
    "Healthcare": [100, 250, 450, 612, 800, 950],
    "Medical Imaging": [50, 120, 200, 250, 350, 400],
    "IoT Health": [30, 80, 150, 200, 250, 300],
    "Smart Cities": [10, 30, 50, 60, 80, 100],
    "Biometrics": [5, 20, 40, 50, 60, 70],
    "Other": [40, 80, 150, 180, 220, 260]
}

# Plotting
fig, ax = plt.subplots(figsize=(12, 6))

for domain, values in data.items():
    ax.plot(years, values, marker='o', label=domain)

ax.set_xlabel("Year")
ax.set_ylabel("Number of Publications")
ax.set_title("Trend of Federated Learning Publications by Domain (2020â€“2025)")
ax.legend()
ax.grid(True)
plt.tight_layout()
plt.show()
```

```python
from scipy.interpolate import make_interp_spline
import numpy as np
import matplotlib.pyplot as plt

years = np.array([2020, 2021, 2022, 2023, 2024, 2025])
data = {
    "Healthcare": [100, 250, 450, 612, 800, 950],
    "Medical Imaging": [50, 120, 200, 250, 350, 400],
}

fig, ax = plt.subplots(figsize=(12, 6))

for domain, values in data.items():
    xnew = np.linspace(years.min(), years.max(), 300)
    spline = make_interp_spline(years, values, k=3)
    ynew = spline(xnew)
    ax.plot(xnew, ynew, label=domain)

ax.set_title("Smoothed Trend Curve (Spline) for Federated Learning")
ax.set_xlabel("Year")
ax.set_ylabel("Publications")
ax.legend()
ax.grid(True)
plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt

years = [2020, 2021, 2022, 2023, 2024, 2025]
data = {
    "Healthcare": [100, 250, 450, 612, 800, 950],
    "Medical Imaging": [50, 120, 200, 250, 350, 400],
    "IoT Health": [30, 80, 150, 200, 250, 300],
}

fig, ax = plt.subplots(figsize=(12, 6))
bottom = [0] * len(years)

for domain, values in data.items():
    ax.fill_between(years, bottom, [sum(x) for x in zip(bottom, values)], label=domain)
    bottom = [sum(x) for x in zip(bottom, values)]

ax.set_title("Stacked Area Chart: FL Publications by Domain")
ax.set_xlabel("Year")
ax.set_ylabel("Number of Publications")
ax.legend()
plt.tight_layout()
plt.show()
```
